# Reading and concatinating the Flight Delay Data

```python
cur_path = '/Users/madhu/Documents/NEU/Fall 2017/ADS Class'
```

```python
#Read in the flight delay and concatinate it
d1 = pd.read_csv(cur_path+'/Data2016.csv', low_memory=False)
d2 = pd.read_csv(cur_path+'/Data2017.csv', low_memory=False)

df = pd.concat([d1,d2])
```

# Data for just the top 10 Airports

```python
#Top Ten airports taken into consideration
airport = ['PHL','DFW','BOS','JFK','IAH','DEN','LGA','ORD','EWR','SFO']
```

```python
#Fintering out data from flight delay to just have the top 10 airport airports
dt_both = df[df.Origin.isin(airport)]
dt_both = dt_both[dt_both.Dest.isin(airport)]
```

# Reading in the Weather Data

```python
df_weather = pd.read_csv(cur_path+'/Top10WeatherData.csv', low_memory=False)
```

```python
df_weather_origin = df_weather[['station', 'Year', 'Month', 'DayofMonth', 'Visibility', 'Wind', 'Precip']]
df_weather_dest = df_weather[['station', 'Year', 'Month', 'DayofMonth', 'Visibility', 'Wind', 'Precip']]
```

```python
df_weather_origin.columns = ['Origin', 'Year', 'Month', 'DayofMonth', 'OriginVisibility', 'OriginWind', 'OriginPrecip']
df_weather_dest.columns = ['Dest', 'Year', 'Month', 'DayofMonth', 'DestinationVisibility', 'DestinationWind', 'DestinationPrecip']
```

# Merging Weather and Flight delay

```python
result = pd.merge(dt_both, df_weather_origin, on=['Origin', 'Year','Month', 'DayofMonth'], how='left')
final_combined = pd.merge(result, df_weather_dest, on=['Dest', 'Year','Month', 'DayofMonth'], how='left')
```

# Dealing with Missing Data

```python
final_combined.OriginVisibility = final_combined.OriginVisibility.fillna((final_combined.OriginVisibility.fillna(method='ffill') +final_combined.OriginVisibility.fillna(method='bfill'))/2)
final_combined.DestinationVisibility = final_combined.DestinationVisibility.fillna((final_combined.DestinationVisibility.fillna(method='ffill') +final_combined.DestinationVisibility.fillna(method='bfill'))/2)
final_combined.OriginWind = final_combined.OriginWind.fillna((final_combined.OriginWind.fillna(method='ffill') +final_combined.OriginWind.fillna(method='bfill'))/2)
final_combined.DestinationWind = final_combined.DestinationWind.fillna((final_combined.DestinationWind.fillna(method='ffill') +final_combined.DestinationWind.fillna(method='bfill'))/2)
final_combined.OriginPrecip = final_combined.OriginPrecip.fillna((final_combined.OriginPrecip.fillna(method='ffill') +final_combined.OriginPrecip.fillna(method='bfill'))/2)
final_combined.DestinationPrecip = final_combined.DestinationPrecip.fillna((final_combined.DestinationPrecip.fillna(method='ffill') +final_combined.DestinationPrecip.fillna(method='bfill'))/2)
```

```python
final_combined.to_csv(cur_path+'/Top10Airports.csv')
```
