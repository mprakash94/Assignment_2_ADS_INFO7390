```python
import pandas as pd
import numpy as np
import csv
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
color = sns.color_palette()
```

```python
cur_path = '/Users/madhu/Documents/NEU/Fall 2017/ADS Class/'
```

```python
d1 = pd.read_csv(open(cur_path+'Airports.csv', 'rU'), encoding = 'utf-8')
```

```python
d1.columns
```

```python
# Origin delay vs origin visibility
plt.subplots(figsize=(18,14))
sns.barplot(x="OriginVisibility", y="ArrDelayMinutes", data=d1)
plt.ylabel('ArrDelayMinutes', fontsize=16)
plt.xlabel('OriginVisibility', fontsize=16)
plt.xticks(rotation='vertical')
plt.title("How DelayMinutes changes with visibility?", fontsize=22)
plt.show()
```

```python
# Origin delay vs origin wind
plt.subplots(figsize=(18,14))
sns.barplot(x="OriginWind", y="ArrDelayMinutes", data=d1)
plt.ylabel('ArrDelayMinutes', fontsize=16)
plt.xlabel('OriginWind', fontsize=16)
plt.xticks(rotation='vertical')
plt.title("How DelayMinutes changes with Wind?", fontsize=22)
plt.show()
```

```python
# Origin delay vs origin precipitation
plt.subplots(figsize=(18,14))
sns.barplot(x="OriginPrecip", y="ArrDelayMinutes", data=d1)
plt.ylabel('ArrDelayMinutes', fontsize=16)
plt.xlabel('OriginPrecip', fontsize=16)
plt.xticks(rotation='vertical')
plt.title("How DelayMinutes changes with Precipitation?", fontsize=22)
plt.show()
```

```python
# Origin delay vs departure visibility
plt.subplots(figsize=(18,14))
sns.barplot(x="DestinationVisibility", y="DepDelayMinutes", data=d1)
plt.ylabel('DepDelayMinutes', fontsize=16)
plt.xlabel('OriginPrecip', fontsize=16)
plt.xticks(rotation='vertical')
plt.title("How departure DelayMinutes changes with Visibility?", fontsize=22)
plt.show()
```

```python
# Origin delay vs departure wind
plt.subplots(figsize=(18,14))
sns.barplot(x="DestinationWind", y="DepDelayMinutes", data=d1)
plt.ylabel('DepDelayMinutes', fontsize=16)
plt.xlabel('DestinationWind', fontsize=16)
plt.xticks(rotation='vertical')
plt.title("How departure DelayMinutes changes with wind?", fontsize=22)
plt.show()
```

```python
# Origin delay vs departure precipitation
plt.subplots(figsize=(18,14))
sns.barplot(x="DestinationPrecip", y="DepDelayMinutes", data=d1)
plt.ylabel('DepDelayMinutes', fontsize=16)
plt.xlabel('DestinationPrecip', fontsize=16)
plt.xticks(rotation='vertical')
plt.title("How departure DelayMinutes changes with Precip?", fontsize=22)
plt.show()
```
