# SORTING

```python
del loan_data['id']
s = loan_data.reset_index()
s['index'] = s.index +1
s=s.rename(columns = {'index':'id'})
```

```python
s.head()
```

```python
f = ['id','application_type','emp_length','dti','annual_inc','dti_joint','annual_inc_joint','grade','sub_grade','int_rate','loan_amnt','issue_d','purpose','addr_state','zip_code']
```

```python
len(f)
```

```python
temp = s[f]
```

```python
temp.isnull().sum()
```

```python
temp = temp[temp.application_type.notnull()]
```

```python
temp.head()
```

## Emp Length

```python
temp['emp_length']=np.where(temp['emp_length']=='10+ years','10',temp['emp_length'])
temp['emp_length']=np.where(temp['emp_length']=='< 1 year','0',temp['emp_length'])
temp['emp_length']=np.where(temp['emp_length']=='1 year','1',temp['emp_length'])
temp['emp_length'] = temp['emp_length'].map(lambda x: x.rstrip('years'))
```

## Zipcode

```python
temp['zip_code'] = temp['zip_code'].astype(str).map(lambda x: x.rstrip('xx'))
```

```python
temp['int_rate'] = temp['int_rate'].astype(str).map(lambda x: x.rstrip('%'))
```

## dti and dti_joint

```python
temp['dti_joint'] = np.where(temp['application_type']=='Individual',0,temp['dti_joint'])
```

## annual_inc_joint

```python
temp['annual_inc_joint'].fillna(0,inplace = True)
```

```python
temp['int_rate']= temp['int_rate'].astype(float)
```
