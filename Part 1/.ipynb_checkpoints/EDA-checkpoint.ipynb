# SORTING

```python
del loan_data['id']
s = loan_data.reset_index()
s['index'] = s.index +1
s=s.rename(columns = {'index':'id'})
```

```python
s.head()
```

```python
f = ['id','application_type','fico_range_low','fico_range_high','emp_length','dti','annual_inc','dti_joint','annual_inc_joint','grade','sub_grade','int_rate','loan_amnt','issue_d','purpose','addr_state','zip_code']
```

```python
len(f)
```

```python
temp = s[f]
```

```python
for i in temp.columns:
    if i not in loan_data.columns:
        print(i)
```

```python
temp.isnull().sum()
```

```python
ann = temp[['dti_joint', 'sub_grade','id']][temp.application_type=='Joint App']
Mode = temp['dti_joint'].mode()[0]
a = list(temp['sub_grade'].unique())

for c in a:
    t = ann[ann.sub_grade == c]
    if( t.dti_joint.isnull().sum()==0):
            break
    else:
            if(t.shape == t[t.isnull().any(axis=1)].shape ):
                t['dti_joint'].fillna(Mode, inplace=True)
            else:
                c = t['dti_joint'].mode()[0]
                t['dti_joint'].fillna(c, inplace=True)
            ann.loc[ann.id.isin(t.id), ['dti_joint']] = t[['dti_joint']]
ann['dti_joint'].fillna(ann['dti_joint'].mode()[0], inplace = True)      
temp.loc[temp.id.isin(ann.id), ['dti_joint']] = ann[['dti_joint']]
```

```python
temp = temp[temp.application_type.notnull()]
```

```python
temp.head()
```

## Emp Length

```python
temp['emp_length']=np.where(temp['emp_length']=='10+ years','10',temp['emp_length'])
temp['emp_length']=np.where(temp['emp_length']=='< 1 year','0',temp['emp_length'])
temp['emp_length']=np.where(temp['emp_length']=='1 year','1',temp['emp_length'])
temp['emp_length'] = temp['emp_length'].map(lambda x: x.rstrip('years'))
```

## Zipcode

```python
temp['zip_code'] = temp['zip_code'].astype(str).map(lambda x: x.rstrip('xx'))
```

```python
temp['int_rate'] = temp['int_rate'].astype(str).map(lambda x: x.rstrip('%'))
```

## Date

```python
temp['Year'] = temp['issue_d'].astype(str).map(lambda x: int(x.split('-')[1]))
temp['Month'] = temp['issue_d'].astype(str).map(lambda x: x.split('-')[0])
```

## dti and dti_joint

```python
temp['dti_joint'] = np.where(temp['application_type']=='Individual',0,temp['dti_joint'])
```
